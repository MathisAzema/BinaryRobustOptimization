```python
# Import des packages n√©cessaires
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# name = "118_syst_JEAS"
# name = "medium"
name = "6bus_JEAS"
df_6_1 = pd.read_excel(f"UnitCommitment_1_{name}.xlsx", sheet_name="results")
df_6_3 = pd.read_excel(f"UnitCommitment_3_{name}.xlsx", sheet_name="results")
df_6_5 = pd.read_excel(f"UnitCommitment_5_{name}.xlsx", sheet_name="results")


# df_list = [df_DCA_l2, df_DCA_l1, df_GRB_l2, df_GRB_l1]
# labels = ["DRO L2 (DCA)", "DRO L1 (DCA)", "DRO L2 (GRB)", "DRO L1 (GRB)"]
```

```python
import ast

res_ccg_L_6_1 =ast.literal_eval(df_6_1['Time per iteration'][0])
res_ccg_L_6_3 =ast.literal_eval(df_6_3['Time per iteration'][0])
res_ccg_L_6_5 =ast.literal_eval(df_6_5['Time per iteration'][0])

res_ccg_big_M_6_1 =ast.literal_eval(df_6_1['Time per iteration'][1])
res_ccg_big_M_6_3 =ast.literal_eval(df_6_3['Time per iteration'][1])
res_ccg_big_M_6_5 =ast.literal_eval(df_6_5['Time per iteration'][1])
```

```python
plt.figure(figsize=(10, 7.5))

res_L = res_ccg_L_6_3
res_big_M = res_ccg_big_M_6_3

plt.plot(res_L, linewidth=4, label="ccg-L")
plt.plot(res_big_M, linewidth=4, label="ccg-big-M")
plt.ylabel("Time (s)", fontsize=30)
plt.xlabel("Iteration", fontsize=30)
plt.tick_params(axis='both', labelsize=30)
plt.legend(fontsize=30, loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.savefig(f"{name}_3.pdf") 
```
